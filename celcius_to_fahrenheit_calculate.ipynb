#CELCİUS TO FAHRENHEİT CALCULATE

##1-import libraries

```python

import numpy as np
```

##2-Set up Training data

```python
celcius_q = np.array([-40,-10,0,8,15,22,38],dtype = "float64")
fahrenheit_a = np.array([-40,14,32,46,59,72,100],dtype = "float64")
```

```python
for i,c in enumerate(celcius_q):
  print("{} degrees Celcius = {} degrees Fahrenhet".format(c,fahrenheit_a[i]))
```

## 3- Create the model

```python
import tensorflow as tf
a = tf.keras.layers.Dense(units=1,input_shape =[1])
```

```python
 model = tf.keras.Sequential([a])
```

```python
model.compile(loss = "mean_squared_error",
              optimizer = tf.keras.optimizers.Adam(0.1))
```

## 4- Train the model

```python
history = model.fit(celcius_q,fahrenheit_a,epochs=500,verbose = False)
print("Fİnished Training the model")
```

## 5-  Display training the statistic

```python
import matplotlib.pyplot as plt
plt.xlabel("Epoch Number")
plt.ylabel("Loss Magnitude")
plt.plot(history.history["loss"])
```

## 6-Predict model

```python
print(model.predict([100]))
```

## 7- looking at the layers weights

```python
print("these are the layers variables  :{}".format(a.get_weights()))
```
